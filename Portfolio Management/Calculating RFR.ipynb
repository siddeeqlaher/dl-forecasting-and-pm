```
rfr = pd.read_csv('/content/drive/My Drive/Portfolio Management/RFR.csv') 
rfr['Date'] = pd.to_datetime(rfr.Date)
rfr.index = rfr['Date'] 
rfr.drop(['Date'], axis=1, inplace=True)

#91 day t-bill value
rfr['Value'] = 100 * (1 - rfr['Rate'] * (91/365))

#Quarterly Risk free rate
rfr['Quarterly RFR'] = (100 / rfr['Value']) - 1

#Weekly Risk free rate
rfr['Weekly RFR'] = (100 / rfr['Value']) ** (7/91) - 1

rfr.head()
```
