```
abg = pd.read_csv('/content/drive/My Drive/Predictions/ABG_Predictions.csv') 
abg.index = abg['Date'] 
abg.drop(['Date'], axis=1, inplace=True)

afe = pd.read_csv('/content/drive/My Drive/Predictions/AFE_Predictions.csv') 
afe.index = afe['Date'] 
afe.drop(['Date'], axis=1, inplace=True)

afx = pd.read_csv('/content/drive/My Drive/Predictions/AFX_Predictions.csv') 
afx.index = afx['Date'] 
afx.drop(['Date'], axis=1, inplace=True)

avi = pd.read_csv('/content/drive/My Drive/Predictions/AVI_Predictions.csv') 
avi.index = avi['Date'] 
avi.drop(['Date'], axis=1, inplace=True)

bvt = pd.read_csv('/content/drive/My Drive/Predictions/BVT_Predictions.csv') 
bvt.index = bvt['Date'] 
bvt.drop(['Date'], axis=1, inplace=True)

fsr = pd.read_csv('/content/drive/My Drive/Predictions/FSR_Predictions.csv') 
fsr.index = fsr['Date'] 
fsr.drop(['Date'], axis=1, inplace=True)

lbh = pd.read_csv('/content/drive/My Drive/Predictions/LBH_Predictions.csv') 
lbh.index = lbh['Date'] 
lbh.drop(['Date'], axis=1, inplace=True)

ned = pd.read_csv('/content/drive/My Drive/Predictions/NED_Predictions.csv') 
ned.index = ned['Date'] 
ned.drop(['Date'], axis=1, inplace=True)

ntc = pd.read_csv('/content/drive/My Drive/Predictions/NTC_Predictions.csv') 
ntc.index = ntc['Date'] 
ntc.drop(['Date'], axis=1, inplace=True)

pik = pd.read_csv('/content/drive/My Drive/Predictions/PIK_Predictions.csv') 
pik.index = pik['Date'] 
pik.drop(['Date'], axis=1, inplace=True)

rlo = pd.read_csv('/content/drive/My Drive/Predictions/RLO_Predictions.csv') 
rlo.index = rlo['Date'] 
rlo.drop(['Date'], axis=1, inplace=True)

sbk = pd.read_csv('/content/drive/My Drive/Predictions/SBK_Predictions.csv') 
sbk.index = sbk['Date'] 
sbk.drop(['Date'], axis=1, inplace=True)

shp = pd.read_csv('/content/drive/My Drive/Predictions/SHP_Predictions.csv') 
shp.index = shp['Date'] 
shp.drop(['Date'], axis=1, inplace=True)

tbs = pd.read_csv('/content/drive/My Drive/Predictions/TBS_Predictions.csv') 
tbs.index = tbs['Date'] 
tbs.drop(['Date'], axis=1, inplace=True)

whl = pd.read_csv('/content/drive/My Drive/Predictions/WHL_Predictions.csv') 
whl.index = whl['Date'] 
whl.drop(['Date'], axis=1, inplace=True)

dfs = [abg, afe, afx, avi, bvt, fsr, lbh, ned, ntc, pik, rlo, sbk, shp, tbs, whl]
df_merged = reduce(lambda left,right: pd.merge(left,right,on=['Date'], how='outer'), dfs)
df_merged = df_merged[3::5] #Keep Mondays only
df_merged = df_merged.dropna()
df_merged.head()
```

```
returns = pd.DataFrame(index=df_merged.index)
returns['ABG Returns'] = (df_merged['ABG Close'].shift(-5) / df_merged['ABG Close']) - 1
returns['AFE Returns'] = (df_merged['AFE Close'].shift(-5) / df_merged['AFE Close']) - 1
returns['AFX Returns'] = (df_merged['AFX Close'].shift(-5) / df_merged['AFX Close']) - 1
returns['AVI Returns'] = (df_merged['AVI Close'].shift(-5) / df_merged['AVI Close']) - 1
returns['BVT Returns'] = (df_merged['BVT Close'].shift(-5) / df_merged['BVT Close']) - 1
returns['FSR Returns'] = (df_merged['FSR Close'].shift(-5) / df_merged['FSR Close']) - 1
returns['LBH Returns'] = (df_merged['LBH Close'].shift(-5) / df_merged['LBH Close']) - 1
returns['NED Returns'] = (df_merged['NED Close'].shift(-5) / df_merged['NED Close']) - 1
returns['NTC Returns'] = (df_merged['NTC Close'].shift(-5) / df_merged['NTC Close']) - 1
returns['PIK Returns'] = (df_merged['PIK Close'].shift(-5) / df_merged['PIK Close']) - 1
returns['RLO Returns'] = (df_merged['RLO Close'].shift(-5) / df_merged['RLO Close']) - 1
returns['SBK Returns'] = (df_merged['SBK Close'].shift(-5) / df_merged['SBK Close']) - 1
returns['SHP Returns'] = (df_merged['SHP Close'].shift(-5) / df_merged['SHP Close']) - 1
returns['TBS Returns'] = (df_merged['TBS Close'].shift(-5) / df_merged['TBS Close']) - 1
returns['WHL Returns'] = (df_merged['WHL Close'].shift(-5) / df_merged['WHL Close']) - 1
returns = returns.dropna()
returns.to_csv(r'/content/drive/My Drive/Portfolio Management/Percentage_Weekly_Returns.csv', index = True, header=True)

returns.head()
```
